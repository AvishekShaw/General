```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

# Create a figure and a set of subplots (axes)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
z = np.sqrt(x)

# Line Plot on ax1
ax1.plot(x, y1, color='royalblue', linestyle='-', marker='o', markersize=5)
ax1.set_title('Line Plot with Custom Style')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.grid(True)

# Scatter Plot on ax2
scatter = ax2.scatter(x, y2, c=z, cmap='viridis', s=100, edgecolor='k')
ax2.set_title('Scatter Plot with Color Map')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.grid(True)

# Add a colorbar for the scatter plot
cbar = plt.colorbar(scatter, ax=ax2, orientation='vertical')
cbar.set_label('Magnitude')

# Annotate specific points
ax1.annotate('Peak', xy=(x[np.argmax(y1)], y1[np.argmax(y1)]), xytext=(x[np.argmax(y1)]+1, y1[np.argmax(y1)]+0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
ax2.annotate('Zero Cross', xy=(x[np.where(np.isclose(y2, 0, atol=0.1))[0][0]], y2[np.where(np.isclose(y2, 0, atol=0.1))[0][0]]),
             xytext=(x[np.where(np.isclose(y2, 0, atol=0.1))[0][0]]+1, y2[np.where(np.isclose(y2, 0, atol=0.1))[0][0]]+0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.dates as mdates

# Sample data for models
np.random.seed(0)
models = {
    'Model': [f'Model {i+1}' for i in range(10)],
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='M'),  # Dates as x-axis
    'MRR': np.random.uniform(0.1, 1.0, 10),  # Mean Reciprocal Rank
    'Recall': np.random.uniform(0.3, 0.8, 10),  # Recall
    'F1 Score': np.random.uniform(0.5, 0.9, 10)  # F1 Score
}

# Convert data to a DataFrame
df = pd.DataFrame(models)

# Create the scatter plot with Seaborn
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Create the scatter plot
scatter = sns.scatterplot(
    data=df, 
    x='Date', 
    y='MRR', 
    hue='Recall', 
    size='F1 Score', 
    sizes=(100, 1000),  # Adjust size scaling
    palette='viridis',  # Color map
    edgecolor='w', 
    alpha=0.7
)

# Add a colorbar
norm = plt.Normalize(df['Recall'].min(), df['Recall'].max())
sm = plt.cm.ScalarMappable(cmap='viridis', norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=scatter, orientation='vertical')
cbar.set_label('Recall')

# Format the x-axis as dates
scatter.xaxis.set_major_locator(mdates.MonthLocator())
scatter.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.gcf().autofmt_xdate()  # Rotate date labels

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Annotate models with improved text placement
for i, row in df.iterrows():
    plt.text(row['Date'], row['MRR'], row['Model'], 
             ha='center', va='center', 
             fontsize=9, weight='bold', color='black')

# Add labels and title with improved font sizes
plt.xlabel('Date', fontsize=12)
plt.ylabel('Mean Reciprocal Rank (MRR)', fontsize=12)
plt.title('Model Performance Over Time', fontsize=14, weight='bold')

# Adjust layout
plt.tight_layout()

plt.show()
```
