```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax=plt.subplots()
x = np.linspace(0,200,100)
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quad')
ax.set_xlabel("numbers")
ax.set_xlim(0,50)
ax.legend()
```

```python
def plotter(ax,x,y,param_dict):
    return(ax.plot(x,y,**param_dict))
```

```python
data1,data2,data3,data4 = np.random.randn(4,100)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
plotter(ax1,data1,data2,{'marker':'x'})
plotter(ax2,data3,data4,{'marker':'o'})
ax2.set_title("Hello")
ax1.set_title("Bye")
ax2.set_ylabel("y of ax2")
ax1.set_ylabel("y of ax1")
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
plotter(ax1,data1,data2,{'marker':'x'})
plotter(ax2,data3,data4,{'marker':'o'})
ax1.set(xlabel="foofoo",xlim=(1,30),title="hello")
ax2.grid()
```

# Histogram


```python
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(100)
```

```python
num_bins=30
fig,ax = plt.subplots()
ax.hist(x,num_bins,density=True)
```
